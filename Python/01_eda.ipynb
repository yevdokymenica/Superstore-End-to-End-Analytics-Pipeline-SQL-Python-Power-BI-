```python
import pandas as pd

# Load dataset and preview first 5 rows

df = pd.read_excel('../data/sample_-_superstore.xls')
df.head()
```

```python
# Check data types, missing values, and summary statistics

df.info()
df.describe()
```

```python
# Remove duplicate rows and verify data shape

df = df.drop_duplicates()
print(f"Number of rows after cleaning: {df.shape[0]}")
```

```python
# Feature Engineering: calculate profit margin and extract date components

df["profit_margin"] = df["Profit"].div(df["Sales"]).fillna(0)
df["order_year"] = df["Order Date"].dt.year
df["order_month"] = df["Order Date"].dt.month
df["order_day"] = df["Order Date"].dt.day

# Preview the newly created features along with Sales and Profit

df[["Sales", "Profit", "profit_margin", "order_year", "order_month"]].head()
```

```python
# Calculate and format core business KPIs: Total Sales, Total Profit, and Overall Margin

total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
overall_margin = total_profit / total_sales

print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Overall Profit Margin: {overall_margin:.2%}")
```

```python
# Analyze total sales by region in descending order

df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
```

```python
# Analyze total profit by category to identify underperforming segments

df.groupby("Category")["Profit"].sum().sort_values()
```

```python
# Identify the top 10 products by total sales

df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
# Analyze the average profit for each individual discount level

df.groupby("Discount")["Profit"].mean()
```

```python
# Segment discounts into High and Low groups to see their impact on average profit
df["discount_group"] = df["Discount"].apply(lambda x: "High" if x >= 0.2 else "Low")
df.groupby("discount_group")["Profit"].mean()
```

```python
# Identify the top 10 customers by total sales volume

df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
# Identify the top 10 most profitable customers

df.groupby("Customer Name")["Profit"].sum().sort_values(ascending=False).head(10)
```

```python
# Analyze total sales trend by year to see annual growth

df.groupby("order_year")["Sales"].sum()
```

```python
# Analyze sales by month to understand seasonality pattern

df.groupby("order_month")["Sales"].sum()
```
